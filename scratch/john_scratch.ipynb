```python
df2 = df.copy()

model_filter = ['gpt-4.1', 'gpt-4.1-mini', 'gpt-4.1-nano', 'gpt-4o', 'gpt-4o-mini', 'claude-sonnet-4-20250514', 'claude-4-opus-20250514']
condition_filter = ['baseline', 'think_about_solution'] #['baseline', 'think_about_solution', 'memorized', 'complex_story', 'python_program', 'generate_random_numbers']
problem_filter = ['train_problem', 'investment_problem', 'boat_problem', 'pipe_problem', 'population_growth', 'decay_problem', 'logarithm_problem', 'compound_interest', 'cosine_problem_1', 'cosine_problem_2']

group_by = ['model_name', 'problem_id', 'condition']

df2 = df2[df2['model_name'].isin(model_filter)]
df2 = df2[df2['condition'].isin(condition_filter)]
df2 = df2[df2['problem_id'].isin(problem_filter)]


model_condition_analysis = df2.groupby(group_by, dropna=False).agg({
    'digits_correct': ['mean', 'count']
}).reset_index()
pd.set_option('display.max_rows', None)

# Flatten the multi-level column index
model_condition_analysis.columns = ['_'.join(col).strip('_') for col in model_condition_analysis.columns]

model_condition_analysis
```

```python

from config.experiments import PHASE1_CONFIG, get_provider_config
from core.data_models import ConditionType
from core.llm_providers import create_provider, LLMProviderError
from engine.experiment_runner import ExperimentRunner

# model_name = 'claude-sonnet-4-20250514'
# model_name = 'gpt-4.1'
model_name = 'gpt-4.1-mini'
config = PHASE1_CONFIG.model_copy()
provider_config = get_provider_config(model_name)
provider = create_provider(provider_config)
runner = ExperimentRunner(provider=provider, writer=None)

problem = next(p for p in config.math_problems if p.id == 'investment_problem') #'boat_problem'
condition =  ConditionType.THINK_ABOUT_SOLUTION
iteration = 0


system_message, user_message = runner._build_prompts(
    problem, condition, iteration, model_name
)

user_message = "1. Think carefully about how you would solve the second question. Write the steps.\n2. A boat travels 45 miles upstream and back. The upstream trip takes 2.3 hours longer than the downstream trip. If the current flows at 3.6 mph, what is the boat's speed in still water?"

# Get completion from LLM
response = await runner.provider.get_completion(
    system_message=system_message,
    user_message=user_message,
    timeout_seconds=config.timeout_seconds
)

print(f"SYSTEM MESSAGE:\n{system_message}")
print(f"USER MESSAGE:\n{user_message}")
print(f"RESPONSE:\n{response}")
print(f"EXPECTED ANSWER:\n{problem.expected_answer}")
print(f"DIGITS CORRECT:\n{runner._parse_response(response, problem, condition)['digits_correct']}")
```
