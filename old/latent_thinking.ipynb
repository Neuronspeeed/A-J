```python
# Ensure the path is set correctly
import sys
import os

# Add the project root to Python path
project_root = '/Users/BrainTech/Documents/AI/A-J'
if project_root not in sys.path:
    sys.path.insert(0, project_root)

# Now import the modules
from config.experiments import MATH_PROBLEMS
from core.data_models import ConditionType

def get_problem_id_from_question(question: str) -> str:
    for problem in MATH_PROBLEMS:
        if problem.question == question:
            return problem.id
    return None

def get_condition_from_filler_question(filler_question) -> str:
    if filler_question == "What is the derivation of the second question?":
        return ConditionType.THINK_ABOUT_SOLUTION.value
    if pd.isna(filler_question):
        return ConditionType.BASELINE.value
    return "other" # other is a placeholder for other conditions
```

```python
# Use absolute path for the CSV file
import os
csv_path = '/Users/BrainTech/Documents/AI/A-J/old/results3.csv'

# Change to the directory containing the CSV if needed
os.chdir('/Users/BrainTech/Documents/AI/A-J/old')

df = pd.read_csv('results3.csv')
df['problem_id'] = df['question'].apply(get_problem_id_from_question)
df['condition'] = df['filler_question'].apply(get_condition_from_filler_question)

df = df[df['condition'].isin(['baseline', 'think_about_solution'])]

groupby = ['model', 'problem_id','condition']
model_filler_analysis = df.groupby(groupby, dropna=False)['num_digits_correct'].agg(['mean', 'count']).reset_index()

# Display the results
print("Average num_digits_correct by model and filler_question:")

# Set pandas display options to show all rows
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

model_filler_analysis
```
